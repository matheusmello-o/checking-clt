```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data_size = 1000
data_samples_size = int(data_size / 2)

# uniform dist
uniform_dist = np.random.uniform(low = 0, high = 1000, size = data_size)

# normal dist
normal_dist = np.random.randn(data_size)

# gamma dist
shape, scale = 2., 2.  # mean=4, std=2*sqrt(2)
gamma_dist = np.random.gamma(shape = shape, scale = scale, size = data_size)
```

# Uniform Distribuition

```python
uniform_dist_mean_100 = []
uniform_dist_mean_1000 = []
uniform_dist_mean_10000 = []

for s in range(100):
    sample_size = np.random.choice(uniform_dist, size = data_samples_size, replace = True)
    uniform_dist_mean_100.append(np.mean(sample_size))

for s in range(1000):
    sample_size = np.random.choice(uniform_dist, size = data_samples_size, replace = True)
    uniform_dist_mean_1000.append(np.mean(sample_size))

for s in range(10000):
    sample_size = np.random.choice(uniform_dist, size = data_samples_size, replace = True)
    uniform_dist_mean_10000.append(np.mean(sample_size))
```

```python
fig, axes = plt.subplots(nrows = 3, ncols = 1, figsize = (10, 8), sharex = True)

sns.histplot(uniform_dist_mean_100, color = 'green', ax = axes[0])
axes[0].set_title('Means Distribuition From an Uniform Distr. with 100 samples')

sns.histplot(uniform_dist_mean_1000, color = 'blue', ax = axes[1])
axes[1].set_title('Means Distribuition From an Uniform Distr. with 1000 samples')

sns.histplot(uniform_dist_mean_10000, color = 'red', ax = axes[2])
axes[2].set_title('Means Distribuition From an Uniform Distr. with 10000 samples')

plt.tight_layout()

plt.show()
```

```python
normal_dist_mean_100 = []
normal_dist_mean_1000 = []
normal_dist_mean_10000 = []

# 100 means sample
for s in range(100):
    sample_size = np.random.choice(normal_dist, size = data_samples_size, replace = True)
    normal_dist_mean_100.append(np.mean(sample_size))

# 1000 means sample
for s in range(1000):
    sample_size = np.random.choice(normal_dist, size = data_samples_size, replace = True)
    normal_dist_mean_1000.append(np.mean(sample_size))

# 10000 means sample
for s in range(10000):
    sample_size = np.random.choice(normal_dist, size = data_samples_size, replace = True)
    normal_dist_mean_10000.append(np.mean(sample_size))
```

```python
fig, axes = plt.subplots(nrows = 3, ncols = 1, figsize = (10, 8), sharex = True)

sns.histplot(normal_dist_mean_100, color = 'green', ax = axes[0])
axes[0].set_title('Means Distribuition From an Normal Distr. with 100 samples')

sns.histplot(normal_dist_mean_1000, color = 'blue', ax = axes[1])
axes[1].set_title('Means Distribuition From an Normal Distr. with 1000 samples')

sns.histplot(normal_dist_mean_10000, color = 'red', ax = axes[2])
axes[2].set_title('Means Distribuition From an Normal Distr. with 10000 samples')

plt.tight_layout()

plt.show()
```
